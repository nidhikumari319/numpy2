other?

Comparing np.mean() and np.average()
Both np.mean() and np.average() compute the average of an array, but they have key differences.

np.mean():

Computes the arithmetic mean.
Syntax: np.mean(a, axis=None, dtype=None, out=None, keepdims=False)
Simple and straightforward.

np.average():

Computes the weighted average if weights are provided.

Syntax: np.average(a, axis=None, weights=None, returned=False)
Use it when you need to specify weights for the elements.
When to use:

Use np.mean() for basic mean calculations.
Use np.average() when working with weighted data.

#3. Describe the methods for reversing a NumPy array along different axes. Provide examples for 1D and 2D
arrays.

ans:

Reversing a NumPy Array
You can reverse a NumPy array using slicing.

1D Array Example:

import numpy as np

arr1D = np.array([1, 2, 3, 4, 5])
reversed_arr1D = arr1D[::-1]
print(reversed_arr1D)  # Output: [5 4 3 2 1]

2D Array Example:

arr2D = np.array([[1, 2, 3], [4, 5, 6]])
reversed_arr2D_row = arr2D[::-1]  # Reverse along the first axis (rows)
reversed_arr2D_col = arr2D[:, ::-1]  # Reverse along the second axis (columns)
print(reversed_arr2D_row)  # Output: [[4 5 6] [1 2 3]]
print(reversed_arr2D_col)  # Output: [[3 2 1] [6 5 4]]

7. Compare vstack() and hstack() functions in NumPy. Provide examples demonstrating their usage and
output.

ans:

Comparing vstack() and hstack()

vstack() and hstack() are used to stack arrays vertically and horizontally, respectively.

Example of vstack():

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6]])
result_vstack = np.vstack((arr1, arr2))
print(result_vstack)
# Output:
# [[1 2]
#  [3 4]
#  [5 6]]
Example of hstack():

result_hstack = np.hstack((arr1, arr2.T))
print(result_hstack)
# Output:
# [[1 2 5]
#  [3 4 6]]

#10. Explain the concepts of vectorization and broadcasting in NumPy. How do they contribute to efficient array operation?

ans:

 Vectorization and Broadcasting
Vectorization: Refers to the ability to perform operations on entire arrays
without explicit loops. This leads to clearer and faster code.

Broadcasting: Allows NumPy to perform operations on arrays of different shapes. It automatically expands the smaller array to match the shape of the larger one.

Example of Vectorization:

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a + b  # Vectorized addition
print(result)  # Output: [5 7 9]

Example of Broadcasting:

a = np.array([[1, 2, 3]])
b = np.array([4, 5, 6])
result = a + b  # b is broadcast to the shape of a
print(result)  # Output: [[5 7 9]]
